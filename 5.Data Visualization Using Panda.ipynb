# Plotting Histograms with Pandas

```python
import pandas as pd
titanic_data = pd.read_csv(r"C:\Users\jaiku\PycharmProjects\Pandas\titanic_data.csv")
```

The first plot we are going to plot is a 
Histogram. There are multiple ways to plot a graph in Pandas. 
The first way is to select the dataframe column by specifying 
the name of the column in square brackets that follows the 

dataframe name and then append the plot name via dot operator. The following script plots a histogram for the Age column of the Titanic dataset using the hist() function. It is important to mention that behind the scenes, the Pandas library makes use of the Matplotlib plotting functions. Therefore, you need to import the Matplotlibâ€™s pyplot module before you can plot Pandas visualizations.

```python
import matplotlib.pyplot as plt 
titanic_data['Age'].hist()
```

```python
titanic_data['Age'].plot(kind='hist')
```

By default, a Pandas histogram divides the data into 10 bins. You can increase or decrease the number of bins by passing an integer value to the bins parameter. The following script plots a histogram for the Age column of the Titanic dataset with 20 bins. 

```python
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set_style('darkgrid')
titanic_data['Age'].hist(bins = 20)
```

```python
titanic_data['Age'].hist(bins = 20, color = 'orange')
```

# Pandas Line Plots

```python
flights_data = sns.load_dataset('flights')
flights_data.head()
```

To plot a line plot, you have to specify the column names for x 
and y axes. If you specify only the column value for the y-axis, 
    the index is used as the x-axis. The following script plots a line plot for the passengers column of the flights data. 

```python
flights_data.plot.line( y='passengers', figsize=(8,6))
```

In the previous examples, we didnâ€™t pass the column name for the x-axis. Letâ€™s see what happens when we specify the year as the column name for the x-axis.

```python
flights_data.plot.line(x ='year', y='passengers', figsize=(8,6), color = 'orange')
```

#  Pandas Bar Plots

To plot bar plots with Pandas, you need a list of categories and a list of values. This list of categories and the list of values must have the same length

Letâ€™s plot a bar plot that shows the average age of male and female passengers. 

To do so, first, we need to calculate the mean age of both male and female passengers traveling in the unfortunate Titanic ship.

The groupby() method of the Pandas dataframe can be used to apply aggregate function with respect to categorical columns. The following script returns the mean values for the ages of male and female passengers for the Titanic ship. 

```python
sex_mean = titanic_data.groupby("Sex")["Age"].mean()
print(sex_mean)
print(type(sex_mean.tolist()))
```

Next, we need to create a new Pandas dataframe with two columns: Gender and Age. 

Then, we can simply use the bar() method to plot a bar plot that displays the average ages of male and female passengers for the Titanic ship.

```python
df = pd.DataFrame({'Gender':['Female', 'Male'], 'Age':sex_mean.tolist()})
ax = df.plot.bar(x='Gender', y='Age', figsize=(8,6))
```

```python
df = pd.DataFrame({'Gender':['Female', 'Male'], 'Age':sex_mean.tolist()})
ax = df.plot.barh(x='Gender', y='Age', figsize=(8,6))
```

```python
df = pd.DataFrame({'Gender':['Female', 'Male'], 'Age':sex_mean.tolist()})
ax = df.plot.bar(x='Gender', y='Age', figsize=(8,6),color='orange')
```

# Pandas Hexagonal Plots

Hexagonal plots are used to plot the density of occurrence 
of values for a specific column. The hexagonal plots will be 
explained with the help of the tips dataset

The following script loads the tips dataset from the Seaborn library and then plots a hexagonal plot that shows values from the total_bill column on the x-axis and values from the tip column on the y-axis.

```python
tips_data = sns.load_dataset('tips')
tips_data.plot.hexbin(x='total_bill', y='tip', gridsize=20, figsize=(8,6))
```

The output shows that most of the time, the tip is between two and four dollars.

```python
# To display a bar plot using the Titanic dataset that shows the average age of passengers who survived vs. those who did not survive, you can follow these steps:
# 
# 1. Import the necessary libraries:


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 2. Load the Titanic dataset:


titanic_data = pd.read_csv('titanic_data.csv')

# 3. Calculate the mean age for survived and not survived passengers:


survived_mean_age = titanic_data[titanic_data['Survived'] == 1]['Age'].mean()
not_survived_mean_age = titanic_data[titanic_data['Survived'] == 0]['Age'].mean()


# 4. Create a DataFrame with the calculated mean ages:


df = pd.DataFrame({
    'Survival': ['Survived', 'Not Survived'],
    'Mean Age': [survived_mean_age, not_survived_mean_age]
})



ax = df.plot.bar(x='Survival', y='Mean Age', figsize=(8, 6))
plt.show()

```
